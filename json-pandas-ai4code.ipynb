```python
import pandas as pd
import os
from pathlib import Path
import json
from tqdm.notebook import tqdm

data_dir = Path('../input/AI4Code')
paths_train = list((data_dir / 'train').glob('*.json'))
paths_test = list((data_dir / 'test').glob('*.json'))


def read_notebook(path):
    return (
        pd.read_json(
            path,
            dtype={'cell_type': 'category', 'source': 'str'})
        .assign(id=path.stem)
        .rename_axis('cell_id')
    )

## reading notebook cells from all json files
notebooks_train = [read_notebook(path) for path in tqdm(paths_train, desc='Train Notebooks')]
notebooks_test = [read_notebook(path) for path in tqdm(paths_test, desc='Test Notebooks')]

def read_notebook(path):
    return (
        pd.read_json(
            path,
            dtype={'cell_type': 'category', 'source': 'str'})
        .assign(id=path.stem)
        .rename_axis('cell_id')
    )

## creating dataframes
train = (
    pd.concat(notebooks_train)
    .set_index('id', append=True)
    .swaplevel()
    .sort_index(level='id', sort_remaining=False)
)
test = (
    pd.concat(notebooks_test)
    .set_index('id', append=True)
    .swaplevel()
    .sort_index(level='id', sort_remaining=False)
)

## saving dataftames to csv files
train.to_csv("train.csv")
test.to_csv("test.csv")

## Reference notebook - https://www.kaggle.com/code/ryanholbrook/getting-started-with-ai4code
```
