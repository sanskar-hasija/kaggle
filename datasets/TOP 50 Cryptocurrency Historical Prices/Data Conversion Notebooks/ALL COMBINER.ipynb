```python
import pandas as pd
import numpy as np
import os
```

```python
import os
names = [] 
for path, subdirs, files in os.walk('D:\datasets\CRYPTO\data\\final\\'):
    
       for filename in files:
            f = os.path.join(path, filename)
            names.append(filename)
```

```python
names
```

```python
count = 0
for i in range(50):
    input_dir = "data/final/" + str(names[i])
    df = pd.read_csv(input_dir)
    count = count + len(df)
print(count)
```

```python
result = pd.DataFrame()
for i in range(50):
    input_dir = "data/final/" + str(names[i])
    df = pd.read_csv(input_dir)
    #del df["SNo"]
    df["Currency_Name"] = str(names[i][:-4])
    cols= ["Currency_Name",'Date', 'Price', 'Open', 'High', 'Low', 'Vol.', 'Change %']
    df = df[cols]
    result = pd.concat([result , df] ,ignore_index = True)
```
