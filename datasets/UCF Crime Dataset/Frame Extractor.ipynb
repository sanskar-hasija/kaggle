```python
import os
import cv2
```

```python
input_path = "data/Videos/"
output_path_train = "data/Dataset/Train/"
output_path_test = "data/Dataset/Test/"
split_file_path = "data/Anomaly_Test.txt"

IMG_HEIGHT = 64
IMG_WIDTH = 64
```

```python
with open(split_file_path) as f:
    mylist = f.read().splitlines() 
test_split_vids = [os.path.join(input_path , mylist[i]) for i in range(len(mylist))] 
```

```python
test_split_vids
```

```python
def image_gen(input_dir , output_path_train, output_path_test ,test_split_vids, img_height , img_width):
    data_files = [input_dir + str(i) for i in os.listdir(input_dir)]
    for data in data_files:
        dir = [data + str("/") + str(i) for i in os.listdir(data)]
        for vid in dir :
            if vid in test_split_vids:
                output_dir = output_path_test + str(vid)[12:len(vid)-4] + "_"
                vidcapture = cv2.VideoCapture(vid)
                success,image = vidcapture.read()
                current_frame = 0 
                frames = []
                while success:
                    image = cv2.resize(image, (img_height , img_width) , interpolation = cv2.INTER_AREA)
                    cv2.imwrite(output_dir + "%d.png" % current_frame, image)
                    if current_frame % 10 == 0 
                        success,image = vidcapture.read()
                    current_frame += 1
                print("Video done(Test) - " ,vid)
            else : 
                output_dir = output_path_train + str(vid)[12:len(vid)-4] + "_"
                vidcapture = cv2.VideoCapture(vid)
                success,image = vidcapture.read()
                current_frame = 0 
                frames = []
                while success:
                    image = cv2.resize(image, (img_height , img_width) , interpolation = cv2.INTER_AREA)
                    cv2.imwrite(output_dir + "%d.png" % current_frame, image)    
                    success,image = vidcapture.read()
                    current_frame += 1
                print("Video done(Train) - " ,vid)
```

```python
%%time 
image_gen(input_path  , 
          output_path_train=output_path_train ,
          output_path_test = output_path_test,
          test_split_vids=test_split_vids,
          img_height=IMG_HEIGHT,
          img_width =IMG_WIDTH)
```
